```python
lst = [1, 2, 3]
lst.append(4)
print(lst)
```

```python
lst = [10, 20, 30]
lst.clear()
print(lst)
```

```python
orig = [1, 2, [3, 4]]
shallow = orig.copy()
shallow.append(5)
shallow[2].append(6)
print("orig:", orig)
print("shallow:", shallow)
```

```python
items = ['a', 'b', 'a', 'c', 'a']
n = items.count('a')
print(n)
```

```python
lst = [1, 2]
lst.extend([3, 4])
lst.extend((5, 6))
print(lst)
```

```python
lst = ['x', 'y', 'z', 'y']
print(lst.index('z'))

try:
    idx = lst.index('z')
except ValueError:
    idx = -1
print("index of 'z':", idx)
```

```python
lst = [0, 1, 3]
lst.insert(2, 2)
print(lst)

lst.insert(0, -1)
print(lst)
```

```python
lst = ['a', 'b', 'c']
last = lst.pop()
print("popped:", last)
print(lst)

second = lst.pop(1)
print("popped index 1:", second)
print(lst)
```

```python
lst = [1, 2, 3, 2, 4]
lst.remove(2)
print(lst)

value = 99
if value in lst:
    lst.remove(value)
else:
    print(f"{value} not in list")
```

```python
lst = [1, 2, 3, 4]
lst.reverse()
print(lst)
```

```python
nums = [3, 1, 4, 2]
nums.sort()
print(nums)

nums.sort(reverse=True)
print(nums)

words = ["apple", "fig", "banana", "date"]
words.sort(key=len)
print(words)
```
